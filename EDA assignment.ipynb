
### Description
The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset contains around 40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Columns are :-

Id : each student given a unique id

Salary : Salary for doing job

DOJ : Date of Joining

DOL : Date of Leaving

Designation : Post of Job

Jobcity : Job Location'

Gender : Male or Female

10percentage : 10th class percent

10board : 10th Board

12graduation : Year of Graduation

12percentage : 12th Percent

12board : 12th Class Board

CollegeID : ID of College

CollegeTier : Tier of College(1,2)

Degree : Degree Obtained

Specialization : Specialization in any field

CollegeGPA : Aggregate GPA at graduation

CollegeCityID : A unique ID to identify the city in which the college is located in

CollegeCityTier : Tier of city where college is present

CollegeState : State where college is present

GraduationYear : Year of Graduation

English : Scores in Language English

Logical : Scores in AMCAT Logical section

Quant : Scores in AMCAT Quantitative section

Domain : Scores in AMCAT’s domain module

ComputerProgramming : Score in AMCAT’s Computer programming section

ElectronicsAndSemicon : Score in AMCAT’s Electronics & Semiconductor Engineering section

ComputerScience : Score in AMCAT’s Computer Science section

MechanicalEngg : Score in AMCAT’s Mechanical Engineering section

ElectricalEngg : Score in AMCAT’s Electrical Engineering section

TelecomEngg : Score in AMCAT’s Telecommunication Engineering section

CivilEngg : Score in AMCAT’s Civil Engineering section

conscientiousness : Standard features of this dimension include high levels of thoughtfulness, good impulse control, and goal-directed behaviors

agreeableness : This personality dimension includes attributes such as trust, altruism, kindness, affection, and other prosocial behaviors.

extraversion : Extraversion (or extroversion) is characterized by excitability, sociability, talkativeness, assertiveness, and high amounts of emotional expressiveness.

neuroticism : Neuroticism is a trait characterized by sadness, moodiness, and emotional instability.

openess_to_experience : This trait features characteristics such as imagination and insight.

## DATA CLEANING

```python
#If a student hasn't opted/attempted an optional test, the same has been mentioned as -1. I will be replacing the
#same with 0.

df.loc[:,['Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']]=\
df.loc[:,['Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']].replace(-1,0)
```

```python
df.head()
```

```python
df['GraduationYear'].value_counts()
```

```python
#on observing the data, some graduation years were mentioned as 0
# checked manually his corresponding 12 yr of passing
df['GraduationYear']=df['GraduationYear'].replace(0,2016) 
```

```python
df['JobCity'].value_counts()
```

```python
# -1 value present in JobCity column, so replace it with NaN values
df['JobCity'] = df['JobCity'].replace('-1',np.nan)
```

```python
df['JobCity']=df['JobCity'].ffill()
```

```python
df['10board'].value_counts()
```

```python
# Replace the 0 value in 10board in np.nan
df['10board'] = df['10board'].replace('0',np.nan)
```

```python
df['12board'].value_counts()
```

```python
# Replace the 0 value in 12board in np.nan
df['12board'] = df['12board'].replace('0',np.nan)
```

# EDA

```python
# Here we make uniform name cbse
df['10board']=df['10board'].replace('central board of secondary education','cbse')

df['12board']=df['12board'].replace('central board of secondary education','cbse')
```

```python
# convert other all board except cbse and icse as state board

df['10board'].replace(to_replace=df['10board'][(df['10board']!='cbse') & (df['10board']!='icse') 
                                                   & (df['10board']!=np.nan)].values,value='state board',inplace=True)


# convert other all board except cbse and icse as state board

df['12board'].replace(to_replace=df['12board'][(df['12board']!='cbse') & (df['12board']!='icse')
                                                   & (df['12board']!=np.nan)].values,value='state board',inplace=True)
```

```python
print(df['10board'].unique())

print(df['12board'].unique())
```

```python
df['Designation'].value_counts()
```

```python
# Here we replace 
df['Designation']=df['Designation'].replace("systems engineer",'system engineer')
```

```python
# Working on specialization column
specialization = df.Specialization.value_counts(ascending = False)
specialization
```

```python
specializationlessthan10 = specialization[specialization<=10]
```

```python
def removespeciallessthan10(value):
    if value in specializationlessthan10:
        return 'other'
    else:
        return value
df.Specialization = df.Specialization.apply(removespeciallessthan10)
df.Specialization.unique()
```

```python
#checking collge GPA since some students have entered a score out of 100, while others have added a score out of 10
df['collegeGPA'].value_counts()
```

```python
#There are values in collegeGPA which are percentage of students, we have to replace it with GPA
df["collegeGPA"]=df[df["collegeGPA"]>1]["collegeGPA"]/9.5
```
